# implementing Linear Regression using pytorch

```python
# important libraries
import pandas as pd
import numpy as np
import torch 
import torch.nn as nn
```

```python
## linearregression
class LinearRegression():
    def __init__(self):
        self.weight= torch.randn(1,1,dtype=torch.float32,requires_grad=True)
        self.bias = torch.randn(1,dtype=torch.float32,requires_grad=True)
    def forward(self,x):
        y = torch.matmul(x,self.weight) + self.bias
        return y
    def loss(self,y,y_pred):
        return ((y_pred - y)**2).mean()
```

```python
# synthic Data

x = torch.randn(100,1)
y = (3*x + 4 ) + 0.01 * torch.randn(100,1) ## resulting y and also add some  randomness
```

```python
model = LinearRegression()
```

```python
epcohs= 100
lr = 0.001
```

```python
y_pred = model.forward(x)
```

```python
y_pred.shape
```

```python
y.shape
```

```python
for epoch in range(1,epcohs+1):
    y_pred = model.forward(x)

    loss = model.loss(y,y_pred)

    if epoch % 10 ==0:
        print(f"epoch : {epoch} :: loss : {loss}")

    ## updating parameters

    

    loss.backward()

    with torch.no_grad():
        model.weight -= lr*model.weight.grad
        model.bias -= lr*model.bias.grad

    model.weight.grad.zero_()
    model.bias.grad.zero_()
```
