## Simple Model train using autograd

```python
#inputs

x = torch.tensor(6.7)
y = torch.tensor(0.0)


#weights
w = torch.tensor(1.0,requires_grad=True)
b = torch.tensor(0.0,requires_grad=True)
```

```python
# loss function binary cross entropy

def binary_cross_entropy(prediction,target):
    e = 1e-8 # to prevent log 0 
    prediction = torch.clamp(prediction,e,1-e) # clamp(x,min,max) boundry that should not above the range min and max 
    return -(target * torch.log(prediction) + (1-target)*torch.log(1-prediction))
```

```python
#forward pass 

z = w*x +b 
y_pred = torch.sigmoid(z)

loss = binary_cross_entropy(y_pred,y)
```

```python
print(loss)
```

```python
# derivates

loss.backward()
```

```python
w.grad
b.grad
```

```python
w = w - 0.001 * w.grad
b = b - 0.001* b.grad
```
